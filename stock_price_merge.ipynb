```python
import pandas as pd
import numpy as np
```

```python
Global_df=pd.read_csv("data/stockprices/GBIME_prices.csv")
NIC_Asia_df=pd.read_csv("data/stockprices/NICA_prices.csv")
Siddhartha_df=pd.read_csv("data/stockprices/SBL_prices.csv")
Nabil_df=pd.read_csv("data/stockprices/NABIL_prices.csv")
Prabhu_df=pd.read_csv("data/stockprices/PRVU_prices.csv")
Agricultural_df=pd.read_csv("data/stockprices/ADBL_prices.csv")
Kumari_df=pd.read_csv("data/stockprices/KBL_prices.csv")
Everest_df=pd.read_csv("data/stockprices/EBL_prices.csv")
Prime_df=pd.read_csv("data/stockprices/PCBL_prices.csv")
Sanima_df=pd.read_csv("data/stockprices/SANIMA_prices.csv")

def list_all_dataframe_names():
    # Get the current global namespace
    global_namespace = globals()
    # Iterate over the global namespace and collect DataFrame names
    dataframe_names = [name for name, obj in global_namespace.items() if isinstance(obj, pd.DataFrame)]

    return dataframe_names

#  get the list of DataFrame names
banks_name = list_all_dataframe_names()
```

```python
banks_name
```

```python
def clean_price_df(df):
    #only keep necessary columns
    df=df[['Symbol','Date','Close','Volume','Percent Change']]
    return df

cleaned_dfs=[]
# Loop through the list and clean each dataframe
for name in banks_name:
    df = globals()[name]  # Access the dataframe by name(string)
    cleaned_df = clean_price_df(df)  # Clean the dataframe
    globals()[name] = cleaned_df  # Optionally, update the original dataframe
    cleaned_dfs.append(cleaned_df)
```

```python
all_banks_prices_df=pd.concat(cleaned_dfs,ignore_index=True)
all_banks_prices_df.set_index('Symbol',inplace=True)
```

```python
all_banks_prices_df['Date']=pd.to_datetime(all_banks_prices_df['Date'],format='mixed')
```

```python
# Function to remove % sign
def remove_percent(x):
    return x.replace('%', '')
# Apply the function to each column using apply
all_banks_prices_df['Percent Change']=all_banks_prices_df['Percent Change'].map(remove_percent)
all_banks_prices_df
```

```python
all_banks_prices_df.to_csv('data/allBanksPrices.csv')
```
