```python
import pandas as pd
import numpy as np
```

```python
#read excel files
nabil_df=pd.read_excel('data/nabil_indicators.xlsx')
nica_df=pd.read_excel('data/nica_indicators.xlsx')
prvu_df=pd.read_excel('data/prvu_indicators.xlsx')
sbl_df=pd.read_excel('data/sbl_indicators.xlsx')
agriculture_df=pd.read_excel('data/agriculture_indicators.xlsx')
everest_df=pd.read_excel('data/everest_indicators.xlsx')

#read csv files
gbime_df=pd.read_csv('data/gbime_indicators.csv')
prime_df=pd.read_csv('data/prime_indicators.csv')
sanima_df=pd.read_csv('data/sanima_indicators.csv')
kumari_df=pd.read_csv('data/kumari_indicators.csv')
```

```python
def format_bank_df(df,selected_row,bank_symbol):
    df=df.iloc[selected_row]
    df=df.set_index('Particulars')
    df=df.T
    df.columns.name=None
    df.index.name='Fiscal Year'
    df.reset_index(inplace=True)
    df['Bank']=bank_symbol
    return df
```

```python
ad_to_bs_fy=fy_year_dict={
    '2022/23':'2079/80',
    '2021/22':'2078/79',
    '2020/21':'2077/78',
    '2019/20':'2076/77',
    '2018/19':'2075/76'
}
```

```python
nabil_df.drop(columns=['S.N.','Indicators'],inplace=True)
nabil_df=format_bank_df(nabil_df,[0,1,3,4,5],'NABIL')
nabil_df
```

```python
new_col_names={'Net Profit / Gross Income':'Net Profit/Gross Income',
             'Earnings Per Share':'Earnings Per Share',
             'Price Earnings Ratio':'Price/Earnings Ratio',
             'Bonus Dividend':'Bonus Dividend',
             'Cash Dividend on Share Capital':'Cash Dividend',
             'Dividend (including bonus) on share capital':'Total Dividend'}
nabil_df.rename(columns=new_col_names,inplace=True)
nabil_df.replace(ad_to_bs_fy,inplace=True)
nabil_df['Total Dividend']=nabil_df['Total Dividend'].fillna(30)
nabil_df['Bonus Dividend']=nabil_df['Total Dividend'].astype('float')-nabil_df['Cash Dividend'].astype('float')
# nabil_df.replace(0.00,0.00001,inplace=True) 
nabil_df
```

### Clean global bank indicators

```python
gbime_df
```

```python
gbime_df.drop(columns=['S.N'],inplace=True)
gbime_df=format_bank_df(gbime_df,[0,2,4,5,6],bank_symbol='GBIME')
gbime_df
```

```python
# Function to remove % sign
def remove_percent(x):
    return x.replace('%', '')
# Apply the function to each column using apply
gbime_df=gbime_df.map(remove_percent)

def clean_fiscal_year(fy_str):
    # Split on space and take the second part
    if pd.notna(fy_str):
        return fy_str.split(' ')[1].split(' (')[0].replace('-','/')
    return fy_str

# Apply the function to the 'fiscal year' column
gbime_df['Fiscal Year'] = gbime_df['Fiscal Year'].apply(clean_fiscal_year)
gbime_df
```

```python
new_col_names={'Net Profit / Gross Income':'Net Profit/Gross Income',
             'Earnings Per Share':'Earnings Per Share',
             'Price Earnings Ratio':'Price/Earnings Ratio',
             'Bonus Dividend':'Bonus Dividend',
             'Cash Dividend on Share Capital':'Cash Dividend',
             'Dividend (including bonus) on Share Capital':'Total Dividend'}
gbime_df.rename(columns=new_col_names,inplace=True)
gbime_df['Bonus Dividend']=gbime_df['Total Dividend'].astype('float')-gbime_df['Cash Dividend'].astype('float')
gbime_df
```

```python
# add 2079/2080 data and remove 2074/75 
gbime_df=gbime_df.loc[0:3]
gbime_df.loc[4] = ['2079/80',23.71 ,22.06,8.43 ,9,8,'GBIME',1] 
gbime_df
```
